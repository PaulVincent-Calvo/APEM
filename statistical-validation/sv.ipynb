## Statistical Analysis

```python
def analyze(metric, model1, model2, name1, name2):
    x = np.array(model1[metric], dtype=float)
    y = np.array(model2[metric], dtype=float)

    mean1, std1 = np.mean(x), np.std(x)
    mean2, std2 = np.mean(y), np.std(y)
    percent_change = ((mean1 - mean2) / abs(mean2)) * 100
    t_stat, p_value = ttest_rel(x, y)

    sig = "✅ Significant" if p_value < 0.05 else "❌ Not significant"
    direction = "▲" if mean1 > mean2 else "▼"

    print(f"  {name1}: {mean1:.4f} ± {std1:.4f}")
    print(f"  {name2}: {mean2:.4f} ± {std2:.4f}")
    print(f"  Change : {direction} {abs(percent_change):.2f}%   p = {p_value:.6f}   {sig}")


def run_comparison(setting_label, apem, baseline, baseline_label, metrics):
    print("=" * 60)
    print(f"  {setting_label}  |  APEM  vs  {baseline_label}")
    print("=" * 60)
    for metric in metrics:
        print(f"\n  [{metric.upper()}]")
        analyze(metric, apem, baseline, "APEM", baseline_label)
    print()
```
