```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold
```

```python
path = "./Datasets/final/scen-training.csv"
```

### Read sampled dataset

```python
df = pd.read_csv(f"{path}")
```

```python
print(df.shape)
```

### Define number of clients

```python
clients = 10
```

### Create stratified folds (preserves label ratios)

```python
skf = StratifiedKFold(n_splits=clients, shuffle=True, random_state=42)
```

### Split dataset between number of clients

```python
client_datasets = []
```

```python
for i, (_, idx) in enumerate(skf.split(df, df['label_encoded'])):
    client_df = df.iloc[idx].reset_index(drop=True)
    client_datasets.append(client_df)
    client_df.to_csv(f"./Datasets/IID/client{i+1}-IID.csv", index=False)
    print(f"Client {i+1} -> {client_df['label_encoded'].value_counts().to_dict()}")
```
