```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Datasets/final/sce-normalized.csv") 
```

```python
train_df, test_df = train_test_split(
    df,
    test_size=0.2,               # 20% for testing
    stratify=df["label_encoded"], # ensure class balance
    random_state=42               # for reproducibility
)
```

```python
train_df.to_csv("Datasets/final/scen-training.csv", index=False)
test_df.to_csv("Datasets/final/scen-testing.csv", index=False)
```
