```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import os
```

```python
df = pd.read_csv("Datasets/final/sample-cleaned-encoded.csv")
```

```python
feature_cols = [c for c in df.columns if c not in ["label", "label_encoded"]]
X = df[feature_cols].apply(pd.to_numeric, errors="coerce").fillna(0)
y = df[["label", "label_encoded"]]

print(f"Number of features: {len(feature_cols)}")
X.head()
```

```python
# Replace infinite values with NaN
X.replace([float("inf"), float("-inf")], pd.NA, inplace=True)

# Fill NaN values with 0
X.fillna(0, inplace=True)

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame and add labels
df_scaled = pd.DataFrame(X_scaled, columns=feature_cols)
df_scaled = pd.concat([df_scaled, y.reset_index(drop=True)], axis=1)

print("Features normalized. Showing first 5 rows:")
df_scaled.head()
```

```python
output_folder = "Datasets/final/"
os.makedirs(output_folder, exist_ok=True)
output_path = os.path.join(output_folder, "sce-normalized.csv")

df_scaled.to_csv(output_path, index=False)
print(f"Normalized dataset saved to: {output_path}")
```
